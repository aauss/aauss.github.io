```python
import matplotlib.pyplot as plt
import numpy as np
import copy
import random
```

```python
new_colors = []

# Some colors I like
colors = [[38,53,52], [59,79,82], [97,135,114], [224,238,184], [231,237,170]] 

for i in range(2000):
    # Create new colors similar to them
    a_color = copy.deepcopy(random.choice(colors))
    channel = random.randint(0,2)
    a_color[channel] += random.randint(-12,12)
    
    # Make these colors accessable to scatter attribute 'color'
    a_color = [channel/255 for channel in a_color]
    a_color.append(1)
    new_colors.append(a_color)
          
```
