```python
import matplotlib.pyplot as plt
import numpy as np
import copy
import random
```

```python
new_colors = []

# Some colors I like
#colors = [[38,53,52], [59,79,82], [97,135,114], [224,238,184], [231,237,170]] 
colors = [[241,241,240],[243,211,197], [39,47,70], [30,35,57], [74,52,50]]

for i in range(2000):
    # Create new colors similar to them
    random_color = copy.deepcopy(random.choice(colors))
    random_channel = random.randint(0,2)
    random_color[random_channel] += random.randint(-12,12)
    
    # Normalize RGB value to 1 to make these colors accessable to plt.scatter attribute 'color'
    normed_random_color = [channel/255 for channel in random_color]
    normed_random_color.append(1)  # Add alpha value for accessability
    new_colors.append(normed_random_color)
          
```

```python
step = 0.04
```

```python
# This function describes how I want the alpha value to decrease with decrease y value
arr = np.arange(start=0, stop=1, step=step)
p = (np.power(range(len(arr)), arr)) 
max_ = max(p)
plt.plot(p)
```

```python
y_values = np.arange(start=0, stop=1, step=step)

from matplotlib.pyplot import figure
figure(figsize=(34, 18))
for i, y in enumerate(y_values):
    x = np.arange(0,40)
    y = [y] * 40
    color = [random.choice(new_colors) for _ in range(40)]
    plt.scatter(x,y, marker="s", 
                color=[random.choice(new_colors) for _ in range(40)],
                alpha=np.power(i,y[0])/max_,
                s=40)
    ax = plt.gca()
    ax.set_facecolor('#002333')
    ax.set_ylim(bottom=0,top=1.01)
    ax.set_xlim(left=0,right=40)
plt.savefig("backgro.png", dpi=200, edgecolor="none")
```
